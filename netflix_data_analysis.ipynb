**Step 1**: import the required libraries

```python
import pandas as pd
```

**Step 2**: Load the Dataset

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
# See first 5 rows
print(df.head())
```

**Step 3** : Basic Exploration

```python
print(df.shape)          # rows, columns
print(df.info())         # datatypes
print(df.isnull().sum()) # check missing values
```

**Step 4**: Answer Key Questions

1. How many Movies vs TV Shows?

```python
print(df['type'].value_counts())
```

  ***Note*** : Netflix has more Movies than TV Shows.

2. What are the most common Genres?

```python
print(df['listed_in'].value_counts().head(10))
```

3. Which Country produces the most content?

```python
print(df['country'].value_counts().head(10))
```

**Note** : The USA produces the most content, followed by India.

4. Content released per Year

```python
print(df['release_year'].value_counts().head(10))
```

```python
df['listed_in'].value_counts().head(10)
```

**Note**: Most content is from Drama, Comedy, and International TV Shows.

```python
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10,5))
```

Note : From that trend → The number of releases has increased significantly after 2015.

**Step 5** : Visualize the Insights

```python
import matplotlib.pyplot as plt

# Movies vs TV Shows
df['type'].value_counts().plot(kind='bar', color=['red', 'blue'])
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
# Top 10 Countries
df['country'].value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Countries with Most Content on Netflix")
plt.show()
```

```python
# Releases over Years
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title("Content Release Trend Over Years")
plt.show()
```
